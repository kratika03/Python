```python
import numpy as np
import pandas as pd
```

```python
order_report=pd.read_excel("Company X - Order Report.xlsx")
pincode_zones=pd.read_excel("Company X - Pincode Zones.xlsx")
sku_master=pd.read_excel("Company X - SKU Master.xlsx")
invoice=pd.read_excel("Courier Company - Invoice.xlsx")
rates=pd.read_excel("Courier Company - Rates.xlsx").T.reset_index()
df=pd.DataFrame()
```

```python
order_report
```

```python
sku_master
```

```python
sku_master.drop_duplicates(inplace= True)
```

```python
df=pd.merge(order_report,sku_master,how='left',on='SKU')
df
```

```python
df['Total Weight'] = df['Order Qty']*df['Weight (g)']
df
```

```python
df=df.groupby('ExternOrderNo').sum().reset_index()
df
```

```python
invoice
```

```python
df=pd.merge(df,invoice,left_on='ExternOrderNo',right_on='Order ID')
df
```

```python
df.drop(['ExternOrderNo','Order Qty','Weight (g)','Zone'],axis=1,inplace=True)
```

```python
df
```

```python
df.columns
```

```python
df = df.iloc[:, [1, 2, 3, 0, 4, 5, 6, 7]]
df.head()
```

```python
pincode_zones.head(1)
```

```python
df=pd.merge(df,pincode_zones,on='Customer Pincode',how='left')
```

```python
df
```

```python
df.drop('Warehouse Pincode_y', axis=1, inplace= True)
df.head()
```

```python
df['Total Weight'] = df['Total Weight']/1000
```

```python
df
```

```python
df.drop_duplicates(inplace= True)
```

```python
df
```

```python
df['Total Weight']=round(df['Total Weight'],2)
```

```python
df
```

```python
bins = [0, 0.5, 1, 1.5, 2, 2.5]
labels = [0.5, 1, 1.5, 2, 2.5]
df['Weight Slab'] = pd.cut(df['Total Weight'], bins=bins, labels=labels, right=False, include_lowest=True)
df
```

```python
df['Additional Weight']=df['Weight Slab'].apply(lambda x: x-0.5)
df
```

```python
forward_fixed = df['Type of Shipment'].apply(lambda x:x.replace(x,'fwd_'))+df['Zone']+"_fixed"
forward_fixed
```

```python
forward_add = df['Type of Shipment'].apply(lambda x:x.replace(x, 'fwd_'))+df['Zone']+"_additional"
forward_add
```

```python
rto_fixed=df['Type of Shipment'].apply(lambda x:x.replace(x,'rto_'))+df['Zone']+'_fixed'
rto_fixed
```

```python
rto_add = df['Type of Shipment'].apply(lambda x:x.replace(x, 'rto_'))+df['Zone']+"_additional"
rto_add
```

```python
df['Fixed Charges']=np.where(df['Type of Shipment']=='Forward charges',forward_fixed,rto_fixed)
df
```

```python
df["Additional Charges"] = np.where(df['Type of Shipment'] == 'Forward charges', forward_add, rto_add)
```

```python
df
```

```python
df=pd.merge(df,rates,how='left',left_on='Fixed Charges',right_on='index')
df
```

```python
df=pd.merge(df,rates,how='left',left_on='Additional Charges',right_on='index')
df
```

```python
df.rename(columns={'0_x': 'Rate_x', '0_y': 'Rate_y'}, inplace=True)
df
```

```python
df['Additional Weight']=df['Additional Weight'].astype('float64')
```

```python
df['Total Additional Charges'] = 2*df['Rate_y']*df['Additional Weight']
df['Total Charges by X'] = df['Rate_x'] + df['Total Additional Charges']
df.head()
```

```python
df.drop(['Fixed Charges', 'Additional Charges', 'index_x', 'Rate_x',
       'index_y', 'Rate_y'], inplace= True, axis=1)
df.head()
```

```python
df['difference']=round(df['Billing Amount (Rs.)']-df['Total Charges by X'],1)
```

```python
df
```

```python
df['Charges'] = np.where(df['difference']>0, "Over Charged", np.where(df['difference']==0, "Same", "Under Charged"))
df.head()
```

```python
df.groupby('Charges').sum()[['Billing Amount (Rs.)','Total Charges by X', 'difference']]
```

```python
df[df['Charges'] == 'Under Charged']
```

```python
with pd.ExcelWriter("Logistics.xlsx") as writer:
    df.to_excel(writer,sheet_name="resultant")
    df[df['Charges']=='']
```

```python
df2=df.groupby('Charges').get_group('Over Charged')
df2
```

```python
df3=df.groupby('Charges').get_group('Under Charged')
df3
```

```python
df4=df.groupby('Charges').get_group('Same')
```

```python
with pd.ExcelWriter("Logistics.xlsx") as writer:
    df.to_excel(writer,sheet_name="resultant")
    df2.to_excel(writer,sheet_name="Over Charged")
    df3.to_excel(writer,sheet_name="Under Charged")
    df4.to_excel(writer,sheet_name="Same")
```
